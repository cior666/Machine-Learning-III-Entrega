```python
import pandas as pd
url = "https://raw.githubusercontent.com/cior666/Machine-Learning-III-Entrega/refs/heads/main/Data%20discovery/spotify_reviews_clean.csv"
df = pd.read_csv(url)
df.head()
df.info()
```

```python
#primero hacemos el tratamiento de minusculas para evitar duplicados por ej Good y good . Adem치s eliminamos los tokens inutiles
#como ser ! ? etc
import string
def procesamiento_basico(text):
    text=text.lower()
    text=text.translate(str.maketrans('', '', string.punctuation))
    return text
df['processed_text']=df['Review'].apply(procesamiento_basico)
```

```python
#ahora vamos a eliminar las stopwords para ello usamos la biblioteca de sklearn. Lo importamos en ingles porq las resenas estan en este idioma
from sklearn.feature_extraction.text import ENGLISH_STOP_WORDS

# Creamos un set de stopwords y preservamos negaciones
stop = set(ENGLISH_STOP_WORDS)
stop -= {"no", "not", "dont", "don't", "nor"}

def eliminar_stopwords(text):
    tokens = text.split()
    tokens = [t for t in tokens if t not in stop]
    return " ".join(tokens)

df['processed_text'] = df['processed_text'].apply(eliminar_stopwords)

#Eliminamos stopwords para reducir ruido y dimensionalidad,
#preservando negaciones como "not" o "don't" ya que aportan
#informaci칩n sem치ntica clave en tareas de an치lisis de sentimiento.
```

```python
#Realizamos un proceso de lemantizacion usando la biblioteca de spaCy
import spacy
nlp=spacy.load("en_core_web_sm")
```

```python
def lematizar(text):
    doc=nlp(text)
    return " ".join([token.lemma_ for token in doc])
df['processed_text'] = df['processed_text'].apply(lematizar)
#Lo que obtenemos haciendo esto es el cambiar a formas mas comunes diversos verbos o adjetivos. Por ejemplo:
#Running por run o worse por bad o better por good
```
