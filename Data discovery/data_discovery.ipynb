```python
import pandas as pd
```

```python
url="https://raw.githubusercontent.com/cior666/Machine-Learning-III-Entrega/refs/heads/main/Data%20discovery/reviews.csv/reviews.csv"
df=pd.read_csv(url)
df.head()
```

```python
#procedemos a ver la info de las columnas que nos interesan
df[['Review', 'Rating']].info()
```

```python
#ahora preparamos las columnas con las que queremos trabajar, como buena prÃ¡ctica eliminamos nulos
df = df[['Review', 'Rating']].dropna()
#tambien eliminamos valores neutros (mitad de puntaje) para evitar ambiguedades
df = df[df['Rating'] != 3]  
#procedemos a hacer la etiqueta binaria
df['sentiment'] = df['Rating'].apply(
    lambda x: 1 if x >= 4 else 0
)
```

```python
#ahora procedemos a copiar las columnas que usamos y generar un nuevo archivo "limpio"
df_clean = df[['Review', 'Rating', 'sentiment']].copy()
df_clean.to_csv(r"C:\Users\silvi\OneDrive\Escritorio\Carrera Data Science\Machine Learning III Entrega\Data discovery\spotify_reviews_clean.csv",index=False)
```
